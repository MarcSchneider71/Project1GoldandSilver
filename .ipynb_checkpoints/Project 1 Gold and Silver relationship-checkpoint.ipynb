```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

"""
here is where we import the modules into our notebook
"""
```

```python
html_page = requests.get('https://www.macrotrends.net/1333/historical-gold-prices-100-year-chart') #Make a get request to retrieve the page
gold = BeautifulSoup(html_page.content, 'html.parser')
```

```python
gold.findAll("tr")
```

```python
yearandprices = gold.findAll("tr")
```

```python
yearandprices
```

```python
yearandprices


"""Once you identify where your information is in the html, use the .text method to pull the value
"""
```

```python
listofhistoricaldata= []
for point in range(len(yearandprices)):
    listofhistoricaldata.append(yearandprices[point].text)
"""we use a for loop to loop through the data to get just the text of all the value points"""
```

```python
listofhistoricaldata
```

```python
listofhistoricaldata[1]
```

```python
data3= []
for a in listofhistoricaldata:
    data3.append(a.split('\n'))
data3
"""here we for a for loop to remove and seperate the values"""
```

```python
PandadataframeGold= pd.DataFrame(data3)
```

```python
dfGoldformatted = PandadataframeGold.drop([x for x in range(53,61)]) #drops the end/blank data points at the end of the dataframe
dfGoldformatted=  dfGoldformatted.drop(0) #drops the first row

dfGoldformatted
```

```python
dfGoldformatted.rename(columns=dfGoldformatted.iloc[0]).drop(dfGoldformatted.index[0])
```
