```python
from bs4 import BeautifulSoup
import requests
# Here, we're just importing both Beautiful Soup and the Requests library
page_linkELA = ("https://coinmarketcap.com/currencies/elastos/historical-data/?start=20130428&end=20190212")
page_linkKCS = ("https://coinmarketcap.com/currencies/kucoin-shares/historical-data/?start=20130428&end=20190212")
page_linkWAN = ("https://coinmarketcap.com/currencies/wanchain/historical-data/?start=20130428&end=20190212")
# this is the url that we've already determined is safe and legal to scrape from.
page_responseELA = requests.get(page_linkELA, timeout=5)
page_responseKCS= requests.get(page_linkKCS, timeout=5)
page_responseWAN= requests.get(page_linkWAN, timeout=5)

# here, we fetch the content from the url, using the requests library
page_contentELA = BeautifulSoup(page_responseELA.content, "html.parser")
page_contentKCS = BeautifulSoup(page_responseKCS.content, "html.parser")
page_contentWAN= BeautifulSoup(page_responseWAN.content, "html.parser")

#we use the html parser to parse the url content and store it in a variable.






"""

Elastos (ELA) bills itself as the 'first completely safe and decentralized infrastructure for the internet'. 
Its vision is to create a blockchain-powered decentralized internet that enables digital rights and secure 
transactions. Elastos uses the Internet as the base-layer infrastructure and separates network communications 
from application computing. It prohibits apps from directly accessing the network so as to prevent malicious attacks. 
Elastos is merged mined with Bitcoin to optimize energy efficiency and leverage the security of the Bitcoin network 
for Elastos applications. Sidechains are also included as an additional layer of security.


Kucoin Shares (KCS) is an exchange based token and native currency of the 
Kucoin crypto exchange. Holders of the token are able to reap the rewards 
from the success of the exchange as 50% of its overall trading fee revenue is shared with 
holders through the tokens. Rewards, received in the form of dividends, are proportionate 
to the amount of Kucoin Shares one holds.

Wanchain (WAN) aims to build a 'super financial market' by connecting distinct digital assets. 
Its key features are cross-chain interoperability, privacy, and smart contract functionality. 
Wanchain connects and facilitates the transfer of value between different blockchains in a 
distributed fashion. Public or private blockchains can integrate with Wanchain to establish connections 
between different ledgers and perform inter-ledger asset transfers. The Wanchain ledger supports not only 
smart contracts, but also token exchange privacy protection. Use cases of the Wanchain include asset transfers, 
loan origination, and credit payments. Wanchain is part of the Enterprise Ethereum Alliance (EEA) and Blockchain 
Interoperability Alliance (BIA). The latter comprises ICON, Aion, and Wanchain and has the shared goal of promoting 
interconnectivity between isolated blockchain networks.


"""
```

```python
ELAcontent = page_contentELA.findAll('tr')
KCScontent = page_contentKCS.findAll('tr')
WANcontent = page_contentWAN.findAll('tr')
```

```python
listofhistoricaldataELA= []
for point in range(len(ELAcontent)):
    listofhistoricaldataELA.append(ELAcontent[point].text)
    
listofhistoricaldataKCS= []
for point in range(len(KCScontent)):
    listofhistoricaldataKCS.append(KCScontent[point].text)
    
listofhistoricaldataWAN= []
for point in range(len(WANcontent)):
    listofhistoricaldataWAN.append(WANcontent[point].text)
```

```python
dataELA= []
for a in listofhistoricaldataELA:
    dataELA.append(a.split('\n'))

dataKCS= []
for a in listofhistoricaldataKCS:
    dataKCS.append(a.split('\n'))
    
dataWAN= []
for a in listofhistoricaldataWAN:
    dataWAN.append(a.split('\n'))
```

```python
import pandas as pd

PandadataframeELA= pd.DataFrame(dataELA)
PandadataframeKCS= pd.DataFrame(dataKCS)
PandadataframeWAN= pd.DataFrame(dataWAN)
```

```python
PandadataframeKCS
```

```python
#drops the end/blank data points at the end of the dataframe
dfELAformatted = PandadataframeELA.drop([x for x in range(377,388)]) 
dfWANformatted = PandadataframeWAN.drop([x for x in range(327,338)])
dfKCSformatted = PandadataframeKCS.drop([x for x in range(477,488)])


#drops the last three columns with no data associated
dfELAformatted = dfELAformatted.drop(dfELAformatted.columns[[8, 9, 10]], axis=1)
dfWANformatted = dfWANformatted.drop(dfWANformatted.columns[[8, 9, 10]], axis=1)
dfKCSformatted = dfKCSformatted.drop(dfKCSformatted.columns[[8, 9, 10]], axis=1)

#renames the columns and drops the top row
dfELAformatted.rename(columns=dfELAformatted.iloc[0]).drop(dfELAformatted.index[0])
dfWANformatted.rename(columns=dfWANformatted.iloc[0]).drop(dfWANformatted.index[0])
dfKCSformatted.rename(columns=dfKCSformatted.iloc[0]).drop(dfKCSformatted.index[0])

dfELAhistorical = dfELAformatted.rename(columns=dfELAformatted.iloc[0]).drop(dfELAformatted.index[0])
dfWANhistorical = dfWANformatted.rename(columns=dfWANformatted.iloc[0]).drop(dfWANformatted.index[0])
dfKCShistorical = dfKCSformatted.rename(columns=dfKCSformatted.iloc[0]).drop(dfKCSformatted.index[0])
```
